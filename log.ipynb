```python
def sign (p1, p2, p3):
    return (p1[0] - p3[0]) * (p2[1] - p3[1]) - (p2[0] - p3[0]) * (p1[1] - p3[1])

def Point_in_Triangle (pt, v1, v2, v3):
    b1 = sign(pt, v1, v2)
    b2 = sign(pt, v2, v3)
    b3 = sign(pt, v3, v1)
    
    has_neg = (b1 < 0) | (b2 < 0) | (b3 < 0)
    has_pos = (b1 > 0) | (b2 > 0) | (b3 > 0)

    return ~(has_neg & has_pos)
```

* Pegando quem est dentro do triangulo formado pelo atacante e as duas traves, 

```python
from dis import dis


goalpost_u = (52.5, 7.)
goalpost_d = (52.5, -7.)

def calculate_distance(v, p1, p2, angle):
    beta = np.arctan((p1[1] - v[1]) / (p1[0] - v[0]))
    phi = np.abs(beta - angle/2)
    delta_x = np.abs(v[0] - goalpost_u[0])
    return delta_x / np.cos(phi)
        
for l in pl:
    gaps = np.zeros(df.shape[0])
    distances = np.zeros(df.shape[0])
    
    pts = len(df) * [[goalpost_u, goalpost_d]]
    points = np.empty(len(pts), dtype=object)
    points[:] = pts
    
    positions_atq = np.stack((df[l + 'x'], df[l + 'y']), axis=-1)
    for r in pr:
        p_in = Point_in_Triangle((df[r + 'x'], df[r + 'y']), (df[l + 'x'], df[l + 'y']), goalpost_u, goalpost_d)
        t = list(zip(df[r + 'x'], df[r + 'y']))
        for i in range(points.shape[0]):
            if p_in.iloc[i]:
                points[i] = points[i] + [t[i]]
        break
    
    # SORT
    for i in range(points.shape[0]):
        points[i] = sorted(points[i], key=lambda tup: tup[1], reverse=True) 
    
    # CALCULO GAP
    for i in range(points.shape[0]):
        gap_angle = None
        distance = None
        for j in range(len(points[i]) - 1):
            # lei cosseno
            l1 = np.linalg.norm(np.array(positions_atq[i]) - np.array(points[i][j]))
            l2 = np.linalg.norm(np.array(positions_atq[i]) - np.array(points[i][j+1]))
            l3 = np.linalg.norm(np.array(points[i][j]) - np.array(points[i][j+1]))
            angle = np.arccos((l1 ** 2 + l2 ** 2 - l3 ** 2) / (2 * l1 * l2))
                  
            if gap_angle is None or gap_angle < angle:
                gap_angle = angle
                distance = calculate_distance(positions_atq[i], points[i][j], points[i][j+1], angle)
        
        gaps[i] = gap_angle
        distances[i] = distance    
    
    # Escrevendo no DF
    df[l + 'gap'] = gaps
    df[l + 'distance'] = distances
```

```python
# quando o passe for sucesso -> pega 1 ciclo antes da prxima ao e pega posio da bola
# quando o passe falhar -> pega 5 ciclos depois da ao e pega posio da bola

df['p_sucess_show_time'] = pd.Series(df.show_time, index=df.index)[(df.action == 'p') & (df.result == 'sucess')]
df['p_sucess_show_time'].fillna(method='bfill', inplace=True)
df['next_pass_sucess_minus_1'] = df['p_sucess_show_time'].where(~((df.action == 'p') & (df.result == 'sucess')), df['p_sucess_show_time'].shift(-1)) - 1
```

* Distncia de Interceptao

```python
def get_position(df, index, col_name):
    col = df.iloc[index][col_name]
    return df.iloc[index][col+'x'], df.iloc[index][col+'y']

p1 = df[['player_id']].apply(lambda id: get_position(df, id.name, 'player_id'), axis=1)#np.stack((df[id+'x'], df[id+'y']), axis=-1))
p2 = list(zip(df['target_pass_x'], df['target_pass_y']))
p3 = df[['enemy_id']].apply(lambda id: get_position(df, id.name, 'enemy_id'), axis=1)#np.stack((df[id+'x'], df[id+'y']), axis=-1))


def sub(x, y):
    return [(a[0]-b[0], a[1]-b[1]) for a, b in zip(x, y)]

norm = np.apply_along_axis(np.linalg.norm, 1, sub(p2, p1))
d = np.abs(np.cross(sub(p2, p1), sub(p1,p3)))/norm

df['intercept_distance'] = d
```

```python
pass_angle = np.arcsin(d/np.linalg.norm(sub(p1, p3)))
df['passing_angle'] = pass_angle
```
